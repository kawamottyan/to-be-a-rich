```python
#インポート文
import numpy as np
import pandas as pd

import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

import lightgbm as lgb
from lightgbm import LGBMRanker

from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import KFold

from sklearn.metrics import confusion_matrix

import optuna

import pydata_google_auth
from google.cloud import bigquery 

import pickle
```

```python
credentials = pydata_google_auth.get_user_credentials(
    ['https://www.googleapis.com/auth/bigquery'],
)
```

```python
client = bigquery.Client(project='keiba-381006', credentials=credentials)
```

```python
horse_query = '''
    SELECT * 
    FROM horse_dataset.main_horse_table 
    LIMIT 1000
    '''
horse_df = client.query(horse_query).to_dataframe()
```

```python
race_query = '''
    SELECT * 
    FROM race_dataset.main_race_table 
    LIMIT 1000
    '''
race_df = client.query(race_query).to_dataframe()
```

```python
#データのマージ
race_df_for_merge = race_df[['race_id',
                            ]]

merged_horse_df = pd.merge(horse_df, race_df_for_merge, on='race_id')
```

```python
#使う特徴量の設定
id = {"race_id",
      "horse_id"
    }

feature = [
           "race_id",
           "horse_id",
           "frame_number",
           "horse_number",
            ]

target = ['rank']

X = merged_horse_df[feature]
y = merged_horse_df[target]
```

```python
X= X.astype(float)
y= y.astype(float)
```

```python
model = lgb.LGBMRegressor() 
model.fit(X, y) 
```

```python
with open('model.pickle', mode='wb') as f:
    pickle.dump(model,f,protocol=2)
```

```python
with open('model.pickle', mode='rb') as f:
    model = pickle.load(f)
```

```python
ans = model.predict(X)
```
